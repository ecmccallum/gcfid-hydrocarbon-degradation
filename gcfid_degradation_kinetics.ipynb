```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Time points (days)
time_days = np.array([0, 3, 6, 9, 12])

# Initial concentration = 50 mg/L for both
initial_conc = 50

# Residual percentages based on paper trends 
hexadecane_percent = np.array([100, 5.2, 4.0, 3.2, 2.76])
phenanthrene_percent = np.array([100, 85, 66.25, 50, 36.56])

# Convert to concentration (mg/L)
hexadecane_conc = initial_conc * (hexadecane_percent / 100)
phenanthrene_conc = initial_conc * (phenanthrene_percent / 100)

# Build DataFrame
df = pd.DataFrame({
    'Time (days)': time_days,
    'Hexadecane (mg/L)': hexadecane_conc,
    'Phenanthrene (mg/L)': phenanthrene_conc
})

print(df)
```

```python
plt.figure(figsize=(8,5))

plt.plot(time_days, hexadecane_conc, 'o-', label='Hexadecane')
plt.plot(time_days, phenanthrene_conc, 's-', label='Phenanthrene')

plt.xlabel('Time (days)')
plt.ylabel('Concentration (mg/L)')
plt.title('Hydrocarbon Degradation Over Time (Synthetic Dataset)')
plt.legend()

plt.show()
```
